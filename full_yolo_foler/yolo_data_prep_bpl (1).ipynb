```python
bbox_dict = {}
```

```python
old_labels_folder = old_folder + '/dataset/labels'
```

```python
for i in os.listdir(old_folder + '/dataset/labels'):
  f = open(old_labels_folder + '/' + i, 'r')
  s = f.readlines()
  bbox_dict[i[:-4]] = s
  f.close()
```

```python
total_data = []
```

```python
for i in os.listdir(old_img_folder):
  # img = cv2.imread(old_img_folder + '/'+ i)
  img_name = i[:-5]
  bbox_img = bbox_dict[img_name]
  total_data.append([old_img_folder + '/'+ i, bbox_img])
```

```python
df = pd.DataFrame(total_data)
```

```python
feed_df = df.sample(frac = 0.95, random_state = 88)
test_df = df[~df.isin(feed_df)].dropna()
train_df = feed_df.sample(frac = 0.85, random_state = 88)
val_df = feed_df[~feed_df.isin(train_df)].dropna()
```

```python
train_df.head()
```

```python
train_images_path = "/content/drive/MyDrive/yolo_bpl_main_folder/new_dataset/images/train"
train_labels_path = "/content/drive/MyDrive/yolo_bpl_main_folder/new_dataset/labels/train"
```

```python
for i in train_df.index:
  info_str = str(train_df[0][i]).split('/')
  img_name = info_str[-1]
  img = cv2.imread(str(train_df[0][i]))
  h, w, c = img.shape
  cv2.imwrite(train_images_path + '/' + img_name, img)

  f = open(train_labels_path + '/' + img_name[:-5] + '.txt', 'w')
  bbox_dict = train_df[1][i]
  f.writelines(bbox_dict)
  f.close()
```
